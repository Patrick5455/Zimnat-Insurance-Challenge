> reorder columns

```python
def movecol(df, cols_to_move=[], ref_col='', place='After'):
    
    cols = df.columns.tolist()    
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])
```

```python
train_set = movecol(train_set, ['age','duration_in_days', 'year', 'month', 'week_of_year'], ref_col='ID')
```

```python
def create_dummies(data, columns):
    
    data_new = data.copy()
    for col in columns:
        data_new = pd.concat([data_new, pd.get_dummies(data_new[col], prefix=col, drop_first=True)],
                               axis=1)
        data_new.drop(labels=col, axis=1, inplace= True)
    
    return data_new
```
